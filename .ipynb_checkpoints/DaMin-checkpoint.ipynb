```python
fitur = df[['StageID','GradeID','SectionID','gender','NationalITy','raisedhands','VisITedResources','ParentAnsweringSurvey','ParentschoolSatisfaction']]
```

```python
label = df['Class']
fitur = df.drop('Class', axis = 1)
fitur = pd.get_dummies(fitur)
```

```python
fitur
```

```python
label
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(fitur, label, test_size = 0.2, random_state = 8)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, confusion_matrix, classification_report
```

```python
def RandomSearch(classifier, random_grid):
    # Definition of the random search
    random_search = RandomizedSearchCV(estimator=classifier,
                                       param_distributions=random_grid,
                                       n_iter=30,
                                       scoring='accuracy',
                                       cv=10, 
                                       verbose=1, 
                                       random_state=8)

    # Fit the random search model
    random_search.fit(X_train, y_train)
    
    print("Parameter terbaik yang dihasilkan dari Random Search : ")
    print(random_search.best_params_) #Print hyperparameter yang telah didapatkan dari RandomizeSearch
    print("")
    print("Rata-rata akurasi dari model dengan hyperparameter yang didapat : ")
    print(random_search.best_score_)
    return random_search.best_estimator_
```

```python
#Decision Tree Classifier
random_grid = {'min_samples_split': [x for x in range(2, 15)], 
              'min_samples_leaf': [x for x in range(1,15)], 
              'max_depth': [x for x in range(2,30)]}
classifier = DecisionTreeClassifier()
bestParamClassifier = RandomSearch(classifier, random_grid)

bestParamClassifier.fit(X_train, y_train)
 
y_train_pred = bestParamClassifier.predict(X_train)
y_test_pred = bestParamClassifier.predict(X_test)

print("Akurasi train data : ",accuracy_score(y_train, y_train_pred))
print("Akurasi test data : ",accuracy_score(y_test, y_test_pred))
# print("Precision train data : ",precision_score(y_train, y_train_pred))
# print("Recall score train data : ",recall_score(y_train, y_train_pred))
# print("F1 Score train data : ",f1_score(y_train, y_train_pred))
print(classification_report(y_test, y_test_pred))
```

```python
#Naive Bayes
random_grid = {'fit_prior' : [True, False],
                   'alpha' : [x*0.1 for x in range(1,15)]}
classifier = MultinomialNB()
bestParamClassifier = RandomSearch(classifier, random_grid)

bestParamClassifier.fit(X_train, y_train)
 
y_train_pred = bestParamClassifier.predict(X_train)
y_test_pred = bestParamClassifier.predict(X_test)

print("Akurasi train data : ",accuracy_score(y_train, y_train_pred))
print("Akurasi test data : ",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python
#SVM
random_grid = {'C': [.0001, .001, .01],
                  'kernel': ['linear'],
                  'gamma': [.0001, .001, .01, .1, 1, 10, 100],
                  'degree': [2, 3, 4, 5]}
classifier =svm.SVC(random_state=8)
bestParamClassifier = RandomSearch(classifier, random_grid)

bestParamClassifier.fit(X_train, y_train)
 
y_train_pred = bestParamClassifier.predict(X_train)
y_test_pred = bestParamClassifier.predict(X_test)

print("Akurasi train data : ",accuracy_score(y_train, y_train_pred))
print("Akurasi test data : ",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python
random_grid = {'n_estimators': [int(x) for x in np.linspace(start = 200, stop = 1000, num = 5)],
               'max_features': ['auto', 'sqrt'],
               'max_depth': [20, 40, 60, 80, 100, None],
               'min_samples_split': [2, 5, 10],
               'min_samples_leaf': [1, 2, 4],
               'bootstrap': [True, False]
}
classifier = RandomForestClassifier(random_state = 8)
bestParamClassifier = RandomSearch(classifier, random_grid)

bestParamClassifier.fit(X_train, y_train)
 
y_train_pred = bestParamClassifier.predict(X_train)
y_test_pred = bestParamClassifier.predict(X_test)

print("Akurasi train data : ",accuracy_score(y_train, y_train_pred))
print("Akurasi test data : ",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```

```python
from sklearn.neural_network import MLPClassifier
# classifier = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(40,20,10), random_state=1)

hidden_layer = []
for i in range(5, 100, 5):
    for j in range(5, 100, 5):
        for k in range(5, 100, 5):
            hidden_layer.append([i,j,k])

random_grid = {'solver' : ['lbfgs'],
               'alpha' : [1e-5],
               'hidden_layer_sizes' : hidden_layer
}
classifier = MLPClassifier(random_state = 8)
bestParamClassifier = RandomSearch(classifier, random_grid)

bestParamClassifier.fit(X_train, y_train)
 
 
y_train_pred = bestParamClassifier.predict(X_train)
y_test_pred = bestParamClassifier.predict(X_test)

print("Akurasi train data : ",accuracy_score(y_train, y_train_pred))
print("Akurasi test data : ",accuracy_score(y_test, y_test_pred))
print(classification_report(y_test, y_test_pred))
```
